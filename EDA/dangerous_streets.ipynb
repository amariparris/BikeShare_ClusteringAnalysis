```python
# Imports 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

# Load Data
data = pd.read_csv('../Data/HeatmapData/danger_routes.csv')
data = data[data['clusters'] != -1]
```

```python
df = data[data.clusters.isin([5,8,9])]
df
```

```python
len(df.start_station_name.unique())
```

```python
top_routes = df.groupby(['start_station_name', 'end_station_name']).size().reset_index(name='count')
top_100_routes = top_routes.sort_values('count', ascending=False).head(100)
top_100_routes
```

```python
import folium
import branca.colormap as cm

# Initializing a map centered around Boston
boston_map = folium.Map(location=[42.3601, -71.0589], zoom_start=13, tiles='cartodbdark_matter')

# Generating a color palette
color_palette = cm.linear.YlGnBu_09.scale(0, top_100_routes['count'].max())
top_100_routes['color'] = top_100_routes['count'].apply(color_palette)

# Adding lines to the map for the top 100 routes with labels and tooltips
for idx, row in top_100_routes.iterrows():
    start_station = df[df['start_station_name'] == row['start_station_name']].iloc[0]
    end_station = df[df['end_station_name'] == row['end_station_name']].iloc[0]
    start_coords = [start_station['start_lat'], start_station['start_lng']]
    end_coords = [end_station['end_lat'], end_station['end_lng']]
    
    # Tooltip and popup
    tooltip = f"Route from {start_station['start_station_name']} to {end_station['end_station_name']}"
    popup = folium.Popup(f"{row['count']} rides", parse_html=True)

    # Drawing the lines with colors based on count
    folium.PolyLine(
        locations=[start_coords, end_coords],
        color=row['color'],
        weight=2.5,
        opacity=0.8,
        tooltip=tooltip,
        popup=popup
    ).add_to(boston_map)

# Adding the color legend
color_palette.caption = 'Number of Rides'
boston_map.add_child(color_palette)

# Saving to an HTML file -- Please open this in a browser for an interactive map, you can hover over the routes for info on their respective stations
boston_map.save('../EDA/dangerous_bike_routes_enhanced.html')
```

```python
# List of Seasonal Rotations

# Map for converting season_ordinal to descriptive names
map_dict = {1: "spring", 2: "summer", 3: "fall", 4: "winter"}

# Step 1: Count the number of rides per station per season
rides_per_station_season = data.groupby(['start_station_name', 'season_ordinal']).size().reset_index(name='number_of_rides')

# Step 2: Calculate the average number of rides for each season across all stations
avg_rides_per_season = rides_per_station_season.groupby('season_ordinal')['number_of_rides'].mean().reset_index(name='seasonal_avg')

# Merge the average back to the original grouped data
rides_per_station_season = pd.merge(rides_per_station_season, avg_rides_per_season, on='season_ordinal')

# Step 3: Filter stations that fall below their season's average number of rides
stations_below_seasonal_avg = rides_per_station_season[rides_per_station_season['number_of_rides'] < rides_per_station_season['seasonal_avg']]

# Step 4: Map 'season_ordinal' to 'Rotational_Season'
stations_below_seasonal_avg['Rotational_Season'] = stations_below_seasonal_avg['season_ordinal'].map(map_dict)

# Prepare to include 'start_station_total_docks'
docks_info = data[['start_station_name', 'start_station_total_docks']].drop_duplicates()

# Step 5: Merge to include 'start_station_total_docks'
result_df = pd.merge(stations_below_seasonal_avg, docks_info, on='start_station_name', how='left')

# Selecting relevant columns for the final DataFrame
final_df = result_df[['start_station_name', 'number_of_rides', 'start_station_total_docks', 'Rotational_Season']].rename(columns={'number_of_rides': 'station_average'})

print (final_df[['start_station_name', 'Rotational_Season']].head(30))
```

```python
# Group by 'Rotational_Season' and then take the first 5 entries for each season
grouped = final_df.groupby('Rotational_Season')

# Use a list comprehension to take the first 5 entries from each group
selected_entries = [group.head(5) for _, group in grouped]

# Concatenate the selected entries back into a DataFrame
selected_df = pd.concat(selected_entries)

# Print the selected DataFrame, showing 'start_station_name' and 'Rotational_Season' columns
print(selected_df[['start_station_name', 'Rotational_Season']])
```
