```python
# Imports
import pandas as pd 
# Data
trips= pd.read_csv('../Data/complete_trips_data_2023.csv', index_col = 0)
stations = pd.read_csv('../Data/BlueBikesData/scoredstations.csv', index_col = 0)
```

```python
trips
```

```python
# Columns to merge for start station
selected_columns = ['Number','danger_score', 'danger_score2', 'Total docks', 'District']

new_column_names = {
    'danger_score': 'start_station_danger_score_.5m', 
    'danger_score2': 'start_station_danger_score_1m', 
    'Total docks': 'start_station_total_docks', 
    'District': 'start_station_district'
}

# Selecting the columns from 'stations'
stations_selected = stations[selected_columns]

# Merging selected start station columns from 'stations' with 'trips'
merged_df = pd.merge(trips, stations_selected, left_on='start_station_id', right_on='Number', how='left')

# Renaming the selected columns
merged_df.rename(columns=new_column_names, inplace=True)

merged_df.drop(columns=['Number'], inplace=True)
```

```python
merged_df
```

```python
merged_df.isnull().sum()
```

```python
new_column_names_end = {
    'danger_score': 'end_station_danger_score_.5m', 
    'danger_score2': 'end_station_danger_score_1m', 
    'Total docks': 'end_station_total_docks', 
    'District': 'end_station_district'
}

# Merging selected end station columns from 'stations' with 'merged_df'
merged_df_2 = pd.merge(merged_df, stations_selected, left_on='end_station_id', right_on='Number', how='left')

# Renaming the selected columns to end station
merged_df_2.rename(columns=new_column_names_end, inplace=True)

merged_df_2.drop(columns=['Number'], inplace=True)
```

```python
merged_df_2 = merged_df_2.dropna()
```
