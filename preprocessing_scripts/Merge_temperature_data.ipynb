```python
import pandas as pd

bike_trips = pd.read_csv('../Data/BlueBikesData/trips_data_2023.csv', index_col = 0)
temp_data = pd.read_csv('../Data/WeatherData/average_temp.csv', index_col = 'Day')

# Convert 'started_at' column to datetime
bike_trips['started_at'] = pd.to_datetime(bike_trips['started_at'])

# Extract just the date
bike_trips['date'] = bike_trips['started_at'].dt.date

bike_trips['date'] = pd.to_datetime(bike_trips['date'])

bike_trips
```

```python
temp_data
```

```python
import numpy as np
import calendar

# Reshape data from wide to long 
temp_df = temp_data.stack().reset_index()
temp_df.columns = ['Day', 'Month', 'AvgTemp']

month_to_number = {'Apr': 4, 'Jul': 7, 'Oct': 10, 'Dec': 12}

temp_df['Day'] = temp_df['Day'].astype(int)

# Map month to number
temp_df['Month'] = temp_df['Month'].map(month_to_number)

# Filter out invalid days for each month
temp_df = temp_df[temp_df.apply(lambda x: x['Day'] <= calendar.monthrange(2023, x['Month'])[1], axis=1)]

# Create dataframe to convert date
date_df = temp_df[['Day', 'Month']].assign(Year=2023)

# convert day and month to datetime column
temp_df['date'] = pd.to_datetime(date_df.rename(columns={'Day': 'day', 'Month': 'month'}))

# Drop day and month columns
temp_df = temp_df.drop(columns = ['Day', 'Month'])

temp_df
```

```python
# Merge temperature data on date column
merged_df = pd.merge(bike_trips, temp_df, on='date', how='left')

merged_df
```
