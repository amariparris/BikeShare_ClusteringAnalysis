```python
# Haversine Function to convert Latlong to miles

def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance in miles between two points 
    on the earth (specified in decimal degrees).
    """
    # Convert decimal degrees to radians 
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula 
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    r = 3956  # Radius of Earth in miles
    return c * r

```

```python
# Danger Score Calculator

def dangerscore(lat, long, accidents_df, radius=0.5, fatal_weight=5):
    """
    Calculate a danger score for a given location based on nearby accidents and their fatality.
    
    Parameters:
    - lat, long: Latitude and longitude of the bikeshare location.
    - accidents_df: DataFrame containing accident data with 'lat', 'long', and 'fatal' columns.
    - radius: Radius in miles to consider for proximity. Default is 0.5 miles.
    - fatal_weight: The weight given to fatal accidents. Default here is 5.
    
    Returns:
    - danger_score: Danger score for the location.
    """
    # Calculate distance from the location to each accident in the dataframe
    distances = accidents_df.apply(lambda row: haversine(lat, long, row['lat'], row['long']), axis=1)
    
    # Filter accidents within the specified radius
    accidents_nearby = accidents_df[distances <= radius]
    
    # Calculate weighted danger score
    danger_score = accidents_nearby.apply(lambda x: 1 + (fatal_weight - 1) * x['fatal'], axis=1).sum()
    
    return danger_score
```

## Danger Score Rationale:

The method for calculating the danger score combines two key elements: the proximity of accidents to a location and the severity (fatality) of those accidents. The rationale behind this approach is as follows:

- Proximity: By focusing on accidents within a specific radius, the method prioritizes local safety conditions, reflecting the immediate risk environment around each bikeshare station. This approach recognizes that accidents occurring closer to a station are more indicative of the safety conditions that bikeshare users might experience.

- Severity (Fatality): Incorporating the severity of accidents through the fatal_weight parameter allows the method to distinguish between the relative impact of fatal and non-fatal accidents. Fatal accidents are given more weight because they represent more severe outcomes, thus indicating higher danger levels.

- Simplicity and Flexibility: This method is relatively straightforward to implement and understand. It allows for adjustments through parameters like the radius for proximity and the weight given to fatal accidents, offering flexibility to tailor the analysis to specific safety concerns or geographic contexts.

At this point, I decided i dont want to make a score that factors in traffic density and time of day, since street infrastucture, traffic light duration, etc. may make more dense areas safer or more open areas more risky. It makes more sense to simply look at accidents and fatalities around a location.

    The danger score is also calculated with historic data, since it may impact customer decisions to use a bike, if the area is dangerous. 

    danger_score default is set based on the average trip distance of our dataset being 1 mile. The part of the avg. journey 'belonging' to the first station would be .5 miles.

**As an alternate approach to the score, danger_score2 calculates based on a 1 mile radius, since this will 'speak' to the danger of the average trip of 1 mile.**

```python
# Add the Danger Score Column to my stations data

stations['danger_score'] = stations.apply(lambda row: dangerscore(row['Latitude'], row['Longitude'],accidents), axis=1)
```

```python
# Add Danger Score 2:

stations['danger_score2'] = stations.apply(lambda row: dangerscore(row['Latitude'], row['Longitude'],accidents,1), axis=1)
```
